```python
class RabbitLeap:
    def __init__(self):
        self.start=['E','E','E','_','W','W','W']
        self.goal=['W','W','W','_','E','E','E']
        
    def goal_test(self, state):
        return state==self.goal

    def gen_moves(self,state):
        moves=[]
        empty=state.index('_')
        n=len(state)
        for i in range(n):
            if(state[i] == 'E'):
                #move right
                if(i+1<n and state[i+1]=='_'):
                    new_state=state[:]
                    new_state[i], new_state[i+1]=new_state[i+1],new_state[i]
                    moves.append((new_state,f"E from {i} to {i+1} "))

                #jump right
                if(i+2<n and state[i+2]=='_' and state[i+1] in ['E','W']):
                    new_state=state[:]
                    new_state[i],new_state[i+2]=new_state[i+2],new_state[i]
                    moves.append((new_state,f"E jumps from {i} to {i+2}"))
            else:
                #move left
                if(i-1>=0 and state[i-1]=='_'):
                    new_state=state[:]
                    new_state[i], new_state[i-1]=new_state[i-1],new_state[i]
                    moves.append((new_state,f"W from {i} to {i-1} "))

                #jump right
                if(i-2>=0 and state[i-2]=='_' and state[i-1] in ['E','W']):
                    new_state=state[:]
                    new_state[i],new_state[i-2]=new_state[i-2],new_state[i]
                    moves.append((new_state,f"W jumps from {i} to {i-2}"))
        return moves

    def bfs(self):
        queue=[(self.start,[])]
        visited=[]
        while(queue):
            current_state,path=queue.pop(0)
            if(current_state in visited):
                continue
            visited.append(current_state)

            if(self.goal_test(current_state)):
                path.append((current_state,"Goal Reached"))
                return path

            for new_state,action in self.gen_moves(current_state):
                new_path=path+[(current_state,action)]
                queue.append((new_state, new_path))
        return None

    def dfs(self):
        stack=[(self.start,[])]
        visited=[]
        while(stack):
            current_state,path=stack.pop()
            if(current_state in visited):
                continue
            visited.append(current_state)

            if(self.goal_test(current_state)):
                path.append((current_state,"Goal Reached"))
                return path

            for new_state,action in (self.gen_moves(current_state)):
                new_path=path+[(current_state,action)]
                stack.append((new_state, new_path))
        return None
            
```

```python
solver=RabbitLeap()
solu_bfs=solver.bfs()
solu_dfs=solver.dfs()
print("solution obtained using bfs")
for step in solu_bfs:
    print(step)
print("solution obtained using dfs")
for step in solu_dfs:
    print(step)
```
